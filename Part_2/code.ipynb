# IMPORTING REQUIRED LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import networkx as nx
from collections import Counter
import seaborn as sbn
```

# LOADING DATASET

```python
lines = open('../data/email-Eu-core.txt', 'r').readlines()
```

# CREATING SET OF NODES AND EDGE LIST

```python
node_set = set()
edge_list = []
for i in range(0, len(lines)):
    line = lines[i].split(' ')
    node_a = int(line[0])
    node_b = int(line[1])
    node_set.add(node_a)
    node_set.add(node_b)
    edge_list.append([node_a, node_b])
```

# CREATING GRAPH

```python
graph = nx.DiGraph()
graph.add_nodes_from(node_set)
graph.add_edges_from(edge_list)
```

# FINDING PAGERANK SCORES

```python
pagerank_score = nx.pagerank(graph)
```

```python
print(f"Page Rank Scores: \n{pagerank_score}")
```

# HUBS AND AUTHORITIES SCORE

```python
hubs_score, authorities_score = nx.hits(graph)
```

```python
print(f"Hubs Score: \n{hubs_score}")
```

```python
print(f"Authorities Score: \n{authorities_score}")
```

# CALCULATING TOP PAGERANK, HUBS, AUTHORITIES BY SCORES

```python
tops_pagerank_by_node_num = sorted(pagerank_score.items(), key=lambda x: x[1], reverse=True)
tops_hubs_by_node_num = sorted(hubs_score.items(), key=lambda x: x[1], reverse=True)
tops_auth_by_node_num = sorted(authorities_score.items(), key=lambda x: x[1], reverse=True)
```

```python
print(f"Top 10 NodeIDs based on PageRank Score: \n{tops_pagerank_by_node_num[:10]}")
```

```python
print(f"Top 10 NodeIDs based on Hub Score: \n{tops_hubs_by_node_num[:10]}")
```

```python
print(f"Top 10 NodeIDs based on Authority Score: \n{tops_auth_by_node_num[:10]}")
```

# PAGERANK DISTRIBUTION AND RUG PLOT

```python
pagerank_distribution = dict(Counter(list(pagerank_score.values())))
```

```python
%matplotlib inline

distribution_sorted = sorted(pagerank_distribution.items(), key= lambda x:x[0])
plt.plot([x[0] for x in distribution_sorted], [y[1] for y in distribution_sorted])
plt.xlabel('PageRank Scores')
plt.ylabel('Frequency')
plt.title('PageRank Distribution')
```

```python
%matplotlib inline

sbn.rugplot(pagerank_score.values())
plt.xlabel('PageRank Scores')
plt.title('PageRank Rug Plot')
```
